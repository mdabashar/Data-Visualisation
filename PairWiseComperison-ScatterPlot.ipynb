```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(6, 46) # each row is a model and each column is a dataset
data.shape
```

```python
def pairwiseComperison(data):
    
    data = data.transpose()
    tm = np.array([row[0] for row in data])
    bm = np.array([max(row[1:]) for row in data])
    
    f, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(right=1)
    ax.set_ylim(top=1)
    ax.fill_between([0, 1], [0, 1], transform=ax.transAxes, alpha=.3)
    ax.scatter(tm, bm)
    ax.plot([0, 1], [0, 1], transform=ax.transAxes)
    plt.xlabel('TSAD-GAN')
    plt.ylabel('Base Lines')

pairwiseComperison(data)
```
