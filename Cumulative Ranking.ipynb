```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(6,46)
data.shape
```

```python
def cumuRanking(data):
    cdata = []
    for row in data:
        crow = []
        c = 0
        row = np.sort(row)
        for col in row:
            c += col
            crow.append(c)
        cdata.append(crow)
    cdata = np.array(cdata)
    f,ax = plt.subplots(figsize=(8,6))
    for crow in cdata:
        ax.plot(range(cdata.shape[1]), crow)
    plt.xlabel('dataset')
    plt.ylabel('cumulative rank')
    plt.legend(['a', 'b', 'c', 'd', 'e', 'f'])
cumuRanking(data)
```
