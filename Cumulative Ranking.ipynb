```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(6,46)
data.shape
```

```python
def cumuRanking(data):
    cdata = []
    sorted_index = np.argsort(np.average(data, axis=0))
    for row in data:
        crow = []
        c = 0
        for idx in sorted_index:
            c += row[idx]
            crow.append(c)
        cdata.append(crow)
    cdata = np.array(cdata)
    f,ax = plt.subplots(figsize=(8,6))
    for crow in cdata:
        ax.plot(range(cdata.shape[1]), crow)
cumuRanking(data)
```
